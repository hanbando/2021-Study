```python
import pandas as pd
df = pd.read_csv("Automobile_data.csv")
df.head()
```

```python
#body-style column의 key를 기준으로 그룹핑
df.groupby('body-style').groups.keys()
```

```python
# body-style -> 'convertible' key로 그룹된 값들 출력
style = df.groupby('body-style')
style.get_group("convertible")
```

```python
#body-style, drive_wheels column의 key를 중심으로 그룹핑
double_grouping = df.groupby(["body-style","drive-wheels"])
double_grouping.first()
```

```python
# dobule_grouping된 데이터 중, normalized-losses column의 최소값
style['normalized-losses'].max()
```

```python
# dobule_grouping된 데이터 중, normalized-losses column의 최소값
style['normalized-losses'].min()
```

```python
# body-stype의 key를 기준으로 평균값
style.mean()
```

```python
# 특정 key에 대한 평균값
style.get_group("convertible").mean()
```
